```python
from fractions import Fraction

def find_sequence_length_greater_than_one(x):
    total = Fraction(0)
    n = 0  # Starting index for summation
    
    while total <= 1:
        total += Fraction(1, (x + 2*n))
        n += 1  # Increment to add the next term in the sequence

    return n  # Return the count of terms needed to exceed the sum of 1

# Calculate the length of the sequence for x values from 1 to 10 as examples
sequence_lengths_for_one = {x: find_sequence_length_greater_than_one(x) for x in range(1, 100)}
sequence_lengths_for_one
```
