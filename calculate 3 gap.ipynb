```python
# Define a function to calculate the sequence x(n+1) = (xn + 1/4) * (2/3)
def calculate_new_sequence(n):
    x = [0]  # Initial value x0 = 0

    # Calculate the sequence up to xn
    for i in range(n):
        x_new = (x[-1] + 1/4) * (2/3)
        x.append(x_new)

    return x

# Calculate the sequence up to x100 as an example
new_sequence_values = calculate_new_sequence(90)
new_sequence_values
```
